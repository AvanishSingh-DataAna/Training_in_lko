```python
import pandas as pd
import numpy as np
```

```python
# concatnate is an operation when you merge two or more data frames together.

df1 = pd.DataFrame({'city':['mumbai','kolkata','delhi'], 'temprature':[43, 38, 41],
                    'humadity':[80,70,65]})
df1
```

```python
df2 = pd.DataFrame({'city':['new york','california','los angeles'], 
                    'temprature':[20, 23, 19],'humadity':[30,25,33]})
df2
```

```python
# now we will apply concat function to join both the data frames.

#  pd.concat([list_of_data_frames_you_want_join])
newdf = pd.concat([df1,df2])
newdf
```

```python
# in the above data frame you can see that index does not changed(or did not adjust according to the data)

# to solve this issue we can pass an attribute ignore_index=True in the concat function.

newdf = pd.concat([df1,df2], ignore_index=True)
newdf
```

```python
# now let's look at how you can create the subset of your data set. or how you can create keys of your data set

newdf = pd.concat([df1,df2], keys=["India","USA"])
```

```python
newdf
```

```python
# now we can find the data set based on the keys using loc(location)

newdf.loc['India']

```

```python
# sometimes their is case when you want to append second data frame with first as column not as rows

# means you just want to append some column to your previous data drame not the whole data frame 

# firstly we will create two new data frames

tdf = pd.DataFrame({'city':['mumbai','delhi','kolkata'], 'temprature':[44, 42, 41]})
tdf
```

```python
wdf = pd.DataFrame({'city':['mumbai','delhi','kolkata'], 'windspeed':[10, 7, 9]})
wdf
```

```python
# now we want to see this windspeed column in the tdf data frame.

# if we follows the previous process of concatnating the two data frames, lets look what happen-:

newdf2 = pd.concat([tdf,wdf])
newdf2
```

```python
xdf = pd.concat([tdf,wdf],axis=1)
xdf
```

```python
# so solve this issue we will pass a argument in the concat function as axis=1.

newdf2 = pd.concat([tdf,wdf['windspeed']],axis=1)
newdf2
```

```python
# some cases you might have missing data in your colums, these cases passing index while concatnating is better option


tdf = pd.DataFrame({'city':['mumbai','delhi','kolkata'], 'temprature':[44, 42, 41]}, index=[0,1,2])

wdf = pd.DataFrame({'city':['delhi','mumbai'], 'windspeed':[10, 7]}, index=[1,0])

newdf3 = pd.concat([tdf,wdf], axis=1)
newdf3
```

```python
# we can also join a dataset with the series

s = pd.Series(['Heat','Driness','cloudy'], name="events")
s
```

```python
# now we will concatnate this series with our data frame as column
newdf4 = pd.concat([tdf,s], axis=1)
newdf4
```
