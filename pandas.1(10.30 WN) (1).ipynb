```python
adf = pd.DataFrame({'city':['new york','chicago','las angeles'],'temprature':[20,22,21]})
bdf = pd.DataFrame({'city':['new york','chicago','Miami'],'humadity':[50,45,46]})
```

```python
# by default it does the inner join
rdf = pd.merge(adf,bdf, on="city")
rdf
```
