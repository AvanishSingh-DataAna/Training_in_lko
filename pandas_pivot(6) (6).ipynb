```python
# now we will learn about the melt

# similer to pivot table melt is also used to transform or reshape data.

edf = pd.read_excel("D:/pandas/humidity_data_melt.xlsx")
edf
```

```python
# now i will apply the melt function on edf.

# syntax-: pd.melt(data_frame_name, id_vars=[list_of_column_you_don't_want_to_make_any_changes])

edf1 = pd.melt(edf, id_vars=['day'])
edf1
```

```python
# if you wnat then you can change that variable and value column names

edf2 = pd.melt(edf, id_vars=["day"], var_name="city", value_name="humidity")
edf2
```

```python
# after applying the melt function it create two new column names variable and value.

# now you can do bunch of filters on that

edf1[edf1['variable']=="london"]
```

```python
edf2[edf2['city']=="london"]
```

```python
edf1[edf1['value']>=33]
```
