```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'city':['new york','chicago','las angeles'],'temprature':[20,22,21]})
df1
```

```python
df2 = pd.DataFrame({'city':['new york','chicago','las angeles'],'humadity':[50,45,46]})
df2
```

```python
# now we will see the merge(). how it is different to concat

# we will how merge can be a option for joining two or more data frames

# syntax-: pd.merge(dataframe1,dataframe2,...., on='column_name_on_which_you_want_to_merge')

df = pd.merge(df1,df2,on='city')
df
```

```python
# one important point about the merge is that it joins only intersection records

# it means if the data is different it will not join that record into the dataframe

# for example-:

df3 = pd.DataFrame({'city':['new york','chicago','manhatten','las vegas'],'temprature':[20,22,21,18]})
df4 = pd.DataFrame({'city':['new york','chicago','las angeles','miami'],'humadity':[50,45,46,55]})
df = pd.merge(df3,df4,on="city")
df
```

```python
df = pd.merge(df3,df4,on="city")
df
```

```python
# as you can see in the results it only merged those records which have common values

# now suppose you want to show all the records not only the intersection one

# you want the union result(outer join)

# so you have to pass how="outer" as an merge attribute.

df = pd.merge(df3,df4, on="city", how="outer")
df
```

```python
# and attribute how="inner"(or inner join/intersection) is default.

# we can do the left and right by providing left and right values to the how attribute

# here left join considers the order in which data set is passed in merge function.

df = pd.merge(df3,df4, on="city", how="left")
df
```

```python
df = pd.merge(df3,df4, on="city", how="right")  
df
```

```python
df = pd.merge(df3,df4, on="city", how="outer")
df
```

```python
# sometime in a big dataset you don't know from where your data has come(either from left/right)

# to know that use an attribute indicator=True(by default it is false)
# it will add one additional column(_merge) to show from where record is coming. 

df = pd.merge(df3,df4, on="city", how="outer", indicator=True)
df
```

```python
# suffexes

# suppose you have two data frames that have same columns in that case you will apply suffexes.

df5 = pd.DataFrame({'city':['new york','chicago','las angeles','las vegas'],
                    'temprature':[20,22,21,18],
                    'humadity':[47,45,46,50]})
df6 = pd.DataFrame({'city':['new york','chicago','las angeles','miami'],
                    'temprature':[20,21,20,18],
                    'humadity':[50,43,49,55]
                   })
newdf = pd.merge(df5, df6, on="city")
newdf
```

```python
newdf = pd.merge(df5, df6, on="temprature", indicator=True)
newdf
```

```python
# as you can for differentiate your dataset it has automatically added suffix _x to the columns.

# but if you want your own suffixes to be added then you can simply specify as an sttribute.

newdf = pd.merge(df5,df6, on="city", suffixes=['_left','_right'])
newdf
```
