```python
import pandas as pd
import sqlalchemy
```

```python
# first thing we will do is to create sqlalchemy engine by using create_engine()

# syntax-: create_engine('python module://username:password@localhost:port_no/database_name)

engine = sqlalchemy.create_engine('mysql+pymysql://root:@localhost:3306/student')
```

```python
# for reading the sql table the method we will use is read-sql_table('table_name','engine_name')

df = pd.read_sql_table('order_detail',engine)
df
```

```python
# if i want to print few columns
df1 = pd.read_sql_table('order_detail',engine, columns=['order_name','order_type','order_amt'])
df1
```

```python
# suppose you want to write some queries here based on mysql

query = '''
select order_detail.order_name, order_detail.order_amt, prod_record.p_name, prod_record.p_amt
from order_detail inner join prod_record on order_detail.order_id=prod_record.p_id
'''
df2 = pd.read_sql_query(query,engine)
df2
```

```python
# how to read the excel data in sql 

# first thing that we will do is to to read that data

newdf = pd.read_csv('D:/pandas/exceldata_for_sql.csv')
newdf
```

```python
# now i want to send this csv data to the sql table

# before that you can rename the column names of your csv data from here is well, using a dictionary

# pd.rename(columns={'name':'user_name','phone':'phone_number'}, inplace=True)

```

```python
# to add the data in my sql table 

newdf.to_sql(name='st_detail',con=engine, index=False, if_exists='replace')
```

```python
# if you don't want to remove previous values and want to insert your data at the end of that table then make id_exist="append"

newdf.to_sql(name='st_detail', con=engine, index=False, if_exists="append")
```

```python
# their is one more method for reading the table

pd.read_sql('st_detail',engine)
```

```python
# and in the same function insted of table name you can supply a query is well
query = '''
select * from st_detail where age>28
'''
pd.read_sql(query,engine)
```
