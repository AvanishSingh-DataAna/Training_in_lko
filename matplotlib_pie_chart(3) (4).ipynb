```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
x = [40,30]
lab = ['Yougster','Adults']
plt.pie(x, labels=lab)


plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# wegprops is a one of the parameter that do some formatting in pie chart.it takes a dictionary.
plt.style.use("fivethirtyeight")
x = [40,60]
lab = ['Yougster','Adults']
plt.pie(x, labels=lab,wedgeprops={'edgecolor':'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
x = [100,90,120,70,80]
lab = ['18year','30year','35year','45year','10year']
clrs = ['red','green','pink','cyan','magenta']
plt.pie(x, labels=lab,wedgeprops={'edgecolor':'black'}, colors=clrs,autopct='%1.1f%%')

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Language Popularity
plt.style.use("fivethirtyeight")
ppl = [59219, 55466, 47544, 36443, 35917, 27097, 23030, 20524, 18523, 7920, 7331, 7201]
lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Ruby', 'Go' ]

plt.pie(ppl, labels=lang,wedgeprops={'edgecolor':'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# use of explode attribute. pass the expload data as 0.1 otherwise 0.

plt.style.use("fivethirtyeight")
ppl = [59219, 55466, 47544, 36443, 35917, 27097, 23030]
lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'C#', 'C++' ]
exp = [0,0,0,0.1,0,0,0]

plt.pie(ppl, labels=lang, explode=exp, wedgeprops={'edgecolor':'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# to see the shadow pass shadow=True and for rotation of chart use startangle parameter 
plt.style.use("fivethirtyeight")
ppl = [59219, 55466, 47544, 36443, 35917, 27097, 23030]
lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'C#', 'C++' ]
exp = [0,0,0,0.1,0,0,0]

plt.pie(ppl, labels=lang, explode=exp, shadow=True, wedgeprops={'edgecolor':'black'},startangle=90)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# to get the percentages of each values takes we should pass auto autopct='%1.1f%%' in plt.pie().
plt.style.use("fivethirtyeight")
ppl = [59219, 55466, 47544, 36443, 35917, 27097, 23030]
lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'C#', 'C++' ]
exp = [0,0,0,0.1,0,0,0]

plt.pie(ppl, labels=lang, explode=exp, shadow=True, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'},startangle=90)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```
