```python
# loading Important Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading inbuilt titanic dataset of seaborn 
df = sns.load_dataset('titanic')
df
```

```python
# plotting a histogram on who column form my dataset
plt.hist(df.who, edgecolor='midnightblue')

plt.show()
```

```python
# managing the y-ticks value so that user can see necessary points 
plt.figure(figsize=(8,9))
y = range(0,600,25)
plt.hist(df.who, edgecolor='midnightblue')
plt.yticks(ticks=y,label=y)
plt.xlabel("Who")
plt.ylabel("No. of People")
plt.grid()
plt.show()
```

```python
# plot for analysing how many people died and how many are alive 
plt.figure(figsize=(8,9))
x = range(0,600,25)
plt.hist(df.alive,color='mediumaquamarine', edgecolor='midnightblue')
plt.yticks(ticks=x,label=x)
plt.xlabel("Alive")
plt.ylabel("No. of People")
plt.show()
```

```python
# here i have used bins to categorize the age data in different groups
# plot of number of people from different age groups in titanic dataset
plt.figure(figsize=(8,7))
b = [0,5,10,15,20,25,30,35,40,45,50,55,60,65,70,75]
x = range(0,76,5)
y = range(0,121,10)
plt.hist(df.age, bins=b, color='mediumorchid', edgecolor='cyan')
plt.xticks(ticks=x, label=x)
plt.yticks(ticks=y,label=y)
plt.xlabel("Age Group",fontsize=20)
plt.ylabel("No. of People",fontsize=20)
plt.grid()
plt.show()
```
