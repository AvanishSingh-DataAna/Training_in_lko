```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# histograms plots bins of our data
# you can divide your data in n number of bins and it will show your plot in those bins.
```

```python
plt.style.use('fivethirtyeight')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.hist(ages,bins=5, edgecolor='black')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

```python
# above data is showing the 5 bins first bin is taking the age<26, second bin is showing the data of ages <32, etc
# you can also manage the bins by passing it as a list

plt.style.use('fivethirtyeight')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
binns = [10, 20, 30, 40, 50, 60]
#y = [0,5,10,15,20,30,40,45,50,55]
plt.hist(ages, bins=binns, edgecolor='black')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

```python
# now this plot is much more understandable then the last one.
# it is showing the numbers of people in my data of different age groups.
# 10-20 age group there are 2 people, 20-30 there are 4 people, 30-40 there are 3 people, 40-50 and 50-60 there are only 1.

# we can also exclude some data if we don't want to add those ages to include in our bin.

# suppose i have exclude the 20-30 age group data.

plt.style.use('fivethirtyeight')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
binns = [20, 30, 40, 50, 60]
plt.hist(ages, bins=binns, edgecolor='black')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

```python
# i have a data which shows the age group of people who have visited the stackoverflow in 2019.
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,6))
data = pd.read_csv('D:/matplotlib/stackover_data_ages.csv')

binns = [10, 20, 30, 40, 50, 60, 70]
plt.hist(data.Age, bins=binns, edgecolor='magenta')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

```python
# to plot the vertical line of median age in my data

plt.style.use('fivethirtyeight')

data = pd.read_csv('D:/matplotlib/stackover_data_ages.csv')
ids = data['No._of_views']
ages = data['Age']

binns = [10, 20, 30, 40, 50, 60, 70]
plt.hist(ages, bins=binns, edgecolor='black')
median_age = 29

plt.axvline(median_age, color='#fc4f30', label='Age Median')

plt.legend()
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```
